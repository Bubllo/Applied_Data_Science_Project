```python
# Importing libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importing Dataset

```python
data = pd.read_csv('dataset.csv')
data
```

```python
data.head()
```

```python
data.info()
```

# visulazation 

## Univariate 

```python
#carrier frequency
plt.scatter(data['Airline'],data['Price'])
plt.xlabel('Airline type')
plt.ylabel('Frequency')
plt.title('Carrier name')
```

```python
#source city 
scity=data['source_city'].value_counts()
scity
```

```python
scityinfo=data['source_city'].unique()
plt.pie(scity,autopct="%.2f",labels=scityinfo)
plt.title("Source city")
```

```python
# destination city
dcity=data['destination_city'].value_counts()
dcity
```

```python
dcityinfo=data['destination_city'].unique()
plt.pie(dcity,autopct="%.2f",labels=dcityinfo)
plt.title("Destination city")
```

```python
# Flight type
flight=data['flight'].value_counts()
flight
```

```python
flightinfo=data['flight'].unique()
plt.pie(flight,autopct="%.2f",labels=flightinfo)
plt.title("Flight model")
```

```python
plt.scatter(data['airline'],data['flight'])
```

## Bivarite

```python
plt.bar(data['stops'],data['duration'])
plt.xlabel('Number of stops')
plt.ylabel('Frequency')
plt.title("Number of stops")
```

# Multivariate Analysis

```python
plt.figure(figsize = (20,10))
sns.heatmap(data.corr(),annot=True)
```
